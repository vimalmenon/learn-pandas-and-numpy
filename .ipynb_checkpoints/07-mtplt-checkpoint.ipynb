```python
import matplotlib.pyplot as plt
```

## Create bar chart

```python
c = ['r', 'y', 'g']
plt.bar([1, 2, 3], [1, 2, 3], color=c)
plt.show()
```

```python
x=["Higest", "Middle", "Low"]
y=[90, 50, 20]
c = ['r', 'y', 'g']
plt.xlabel("Medium", fontsize=20)
plt.ylabel("Score", fontsize=20)
plt.bar(x, y, width=0.2, color=c, edgecolor="b", linewidth=2, linestyle=":", alpha=0.5)
plt.show()
```

```python
x=["Higest", "Middle", "Low"]
y=[90, 50, 20]
c = ['r', 'y', 'g']
plt.xlabel("Medium", fontsize=20)
plt.ylabel("Score", fontsize=20)
plt.bar(x, y, width=0.2, label="test")
plt.legend()
plt.show()
```

```python
x=["Higest", "Middle", "Low"]
y=[90, 50, 20]
z=[100, 80, 50]
plt.xlabel("Medium", fontsize=20)
plt.ylabel("Score", fontsize=20)
plt.barh(x, y, height=0.2)
plt.show()
```
